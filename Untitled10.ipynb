```python
import numpy as np
x=np.arange(0,10,1)
y=x**2+2*x*np.pi+5
plt.plot(x,y,"r--")
plt.title("График y=x**2+2*x*np.pi+5")
plt.xlabel("Ось x")
plt.ylabel("Ось y")
plt.grid()
plt.show()
```
