```python
# VEHICLE CLASS
# Бүх тээврийн хэрэгслийн үндсэн класс
class Vehicle:
  # Vehicle тээврийн хэрэгсэл
    # Constructor
    # Обьект үүсэхэд ажиллана
    def __init__(self, brand, model, year):
        # Обьектын өгөгдөл
        self.brand = brand
        self.model = model
        self.year = year
        # Машин асаалттай эсэх
        self.is_running = False
    # Машин асаах
    def start(self):
        self.is_running = True
        print(self.brand, self.model, "аслаа.")
    # Машин унтраах
    def stop(self):
        self.is_running = False
        print(self.brand, self.model, "унтарлаа.")
    # Машины мэдээлэл
    def info(self):
        print("----- INFO -----")
        print("Brand :", self.brand)
        print("Model :", self.model)
        print("Year  :", self.year)
        if self.is_running:
            print("Status: Асаалттай")
        else:
            print("Status: Унтарсан")
# CAR CLASS
# Vehicle-ээс удамшиж байна
class Car(Vehicle):
    # fuel_type нэмсэн constructor
    def __init__(self,
                 brand,
                 model,
                 year,
                 fuel_type):
        # Parent constructor
        super().__init__(brand, model, year)
        # Шинэ attribute
        self.fuel_type = fuel_type
    # Машин жолоодох
    def drive(self):

        if self.is_running:
            print(self.brand, self.model, "явж байна.")
        else:
            print("Эхлээд асаана уу.")
# ==========================================
# ELECTRIC CAR CLASS
# Car-оос удамшиж байна
# ==========================================
class ElectricCar(Car):
    # battery_capacity нэмсэн constructor
    def __init__(self,
                 brand,
                 model,
                 year,
                 fuel_type,
                 battery_capacity):
        # Parent constructor
        super().__init__(brand,
                         model,
                         year,
                         fuel_type)
        # Battery хэмжээ
        self.battery_capacity = battery_capacity
    # Цэнэглэх method
    def charge(self):
        print(self.brand, self.model, "цэнэглэж байна.")
    # Method overriding
    # start()-ийг өөрчилж байна
    def start(self):

        self.is_running = True

        print(self.brand,
              self.model,
              "чимээгүй аслаа.")
# OBJECT ҮҮСГЭХ
car1 = Car("Toyota",
           "Camry",
           2020,
           "Petrol")
tesla1 = ElectricCar("Tesla",
                     "Model S",
                     2024,
                     "Electric",
                     "100 kWh")
# CAR TEST
print("\n=== CAR TEST ===\n")
car1.info()
print()
car1.start()
print()
car1.drive()
print()
car1.stop()
# ==========================================
# TESLA TEST
# ==========================================
print("\n=== TESLA TEST ===\n")
tesla1.info()
print()
tesla1.start()
print()
tesla1.drive()
print()
tesla1.charge()
print()
tesla1.stop()
```

#Pandas сан

```python
# Pandas санг импорт хийх
import pandas as pd
# -----------------------------
# 1. DataFrame үүсгэх
# -----------------------------

# Python dictionary ашиглан өгөгдөл үүсгэнэ
data = {
    "Нэр": ["Ариун", "Болд", "Сараа", "Тэмүүжин"],
    "Нас": [20, 21, 19, 22],
    "Оноо": [85, 92, 78, 95]
}

# DataFrame бол хүснэгтэн өгөгдөл
df = pd.DataFrame(data)

print("=== Бүх өгөгдөл ===")
print(df)


# -----------------------------
# 2. Эхний мөрүүдийг харах
# -----------------------------

print("\n=== Эхний 2 мөр ===")
print(df.head(2))


# -----------------------------
# 3. Нэг багана сонгох
# -----------------------------

print("\n=== Нэр багана ===")
print(df["Нэр"])


# -----------------------------
# 4. Нөхцөлөөр шүүх (filter)
# -----------------------------

# Оноо нь 90-ээс их хүмүүсийг авах
high_score = df[df["Оноо"] > 90]

print("\n=== 90-ээс дээш оноотой хүмүүс ===")
print(high_score)


# -----------------------------
# 5. Шинэ багана нэмэх
# -----------------------------

# Тэнцсэн эсэхийг шинэ багана болгоно
df["Тэнцсэн"] = df["Оноо"] >= 80

print("\n=== Шинэ багана нэмсний дараа ===")
print(df)


# -----------------------------
# 6. Дундаж тооцоолох
# -----------------------------

average_score = df["Оноо"].mean()

print("\n=== Дундаж оноо ===")
print(average_score)


# -----------------------------
# 7. Хамгийн өндөр оноо
# -----------------------------

max_score = df["Оноо"].max()

print("\n=== Хамгийн өндөр оноо ===")
print(max_score)


# -----------------------------
# 8. Мөрөөр хандах
# -----------------------------

print("\n=== 1-р индексийн мэдээлэл ===")
print(df.loc[1])


# -----------------------------
# 9. CSV файл хадгалах
# -----------------------------

# index=False гэвэл мөрийн дугаар хадгалахгүй
df.to_csv("students.csv", index=False)

print("\nCSV файл амжилттай хадгалагдлаа!")
```
