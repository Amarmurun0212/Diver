```python
x=np.linspace(0,4*np.pi,100)
plt.plot(x,np.sin(x),c="r")
plt.plot(x,np.cos(x),c="g")
plt.title("function sin(x) and cos(x)")
plt.grid()
plt.legend(fontsize=20, loc=0)
plt.show()
```
