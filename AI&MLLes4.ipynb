```python
# input()  -> Хэрэглэгчээс текст хэлбэрээр утга авна
# split()  -> Текстийг зайгаар тасалж list болгоно
# map()    -> List доторх string утгуудыг int төрөл рүү хөрвүүлнэ
# list()   -> map object-ийг жинхэнэ list болгоно

# Жишээ:
# Input: 1 2 3 4

arr = list(map(int, input("Тоонууд оруулна уу: ").split()))

# Дээрх мөр алхам алхмаар ингэж ажиллана:

# 1. input()
# Хэрэглэгчээс:
# "1 2 3 4"
# гэсэн текст авна

# 2. split()
# "1 2 3 4".split()
# =>
# ['1', '2', '3', '4']
#
# Өөрөөр хэлбэл зайгаар салгаж list болгодог

# 3. map(int, ...)
# map(int, ['1', '2', '3', '4'])
# =>
# 1, 2, 3, 4
#
# String утгуудыг integer (int) төрөл рүү хөрвүүлдэг

# 4. list()
# list(...) ашигласнаар:
# [1, 2, 3, 4]
# гэсэн list болно


# -----------------------------------------
# Тоо бүрийн давтамжийг хадгалах dictionary
# -----------------------------------------
count_dict = {}

# arr list-ийн элемент бүрээр давталт хийнэ
for el in arr:

    # Хэрэв тухайн тоо dictionary-д өмнө нь байгаа бол
    if el in count_dict:

        # Утгыг 1-ээр нэмэгдүүлнэ
        count_dict[el] += 1

    else:
        # Байхгүй бол шинэ key үүсгээд 1 гэж онооно
        count_dict[el] = 1

# -----------------------------------------
# Үр дүн хэвлэх
# -----------------------------------------
print("Давтамжийн үр дүн:")
print(count_dict)
```

```python
# Нийлбэр хадгалах хувьсагч
total = 0

# range(1, 11)
# 1-ээс 10 хүртэл давтана
for i in range(1, 11):

    # total дээр i-ийн утгыг нэмнэ
    total += i

    # Явцын мэдээлэл хэвлэх
    print("Одоогийн тоо:", i)
    print("Нийлбэр:", total)

# Эцсийн үр дүн
print("Эцсийн нийлбэр =", total)
```

```python
# Нийлбэр хадгалах хувьсагч
total = 0

# Эхлэх утга
i = 1

# i нь 10-аас бага эсвэл тэнцүү байх хүртэл давтана
while i <= 10:

    # total дээр i-ийн утгыг нэмнэ
    total += i

    # Явцын мэдээлэл хэвлэх
    print("Одоогийн тоо:", i)
    print("Нийлбэр:", total)

    # i-ийн утгыг 1-ээр нэмэгдүүлнэ
    i += 1

# Эцсийн үр дүн
print("Эцсийн нийлбэр =", total)
```

# давталтан дотор хийгдэх үйлдлүүд

```python
'''Pass
Бодлого:
1-ээс 5 хүртэл тоо хэвлэхдээ 3 дээр юу ч хийхгүй алгас.'''
# 1-ээс 5 хүртэл давталт
for i in range(1, 6):
    # Хэрэв i = 3 бол
    if i == 3:
        # pass = юу ч хийхгүй
        pass
    else:
    # Тоог хэвлэх
      print(i)
'''
Түр хоосон үлдээх команд
Код алдаа өгөхөөс сэргийлдэг
“Одоохондоо юу ч хийхгүй” гэсэн утгатай
'''
```

```python
'''
break
Давталтыг шууд зогсоодог
break ажиллахад loop цааш үргэлжлэхгүй
Бодлого:
1-ээс 10 хүртэл хэвлэхдээ 5 дээр давталтыг бүрэн зогсоо.'''
# 1-ээс 10 хүртэл давталт
for i in range(1, 11):
    # Хэрэв i = 5 бол
    if i == 5:
        # Давталтыг бүрэн зогсооно
        break
    # Тоо хэвлэх
    print(i)
```

```python
'''
Бодлого:
1-ээс 5 хүртэл хэвлэхдээ 3-ыг алгас.
'''
# 1-ээс 5 хүртэл давталт
for i in range(1, 6):
    # Хэрэв i = 3 бол
    if i == 3:
        # Энэ алхмыг алгасаад дараагийн давталт руу орно
        continue
    # Тоо хэвлэх
    print(i)
'''
continue
Одоогийн давталтыг алгасна
Дараагийн давталт руу шууд шилжинэ
'''
```

```python
# else
'''
else ашигласан давталт
Бодлого:
1-ээс 5 хүртэл тоо хэвлээд давталт амжилттай дуусвал message гарга.
'''
# 1-ээс 5 хүртэл давталт
for i in range(1, 6):
    # Тоо хэвлэх
    print(i)
# Давталт хэвийн дууссан бол ажиллана
else:
    print("Давталт амжилттай дууслаа")
# Давталт break-гүй дууссан үед ажиллана
```

```python
'''
break + else хамт ашигласан жишээ
Бодлого:
Тоо хайж олоод байвал stop хий.
'''
# Жагсаалт
numbers = [1, 3, 5, 7, 9]
# Хайх утга
target = 5
# Давталт
for num in numbers:
    # target олдвол
    if num == target:
        print("Тоо олдлоо")
        # Давталтыг зогсооно
        break
# break ажиллаагүй үед else ажиллана
else:
    print("Тоо олдсонгүй")
```

Команд	Үүрэг
pass	= Юу ч хийхгүй
break	= Давталтыг бүрэн зогсооно
continue	= Одоогийн алхмыг алгасна
else	= Давталт хэвийн дуусвал ажиллана
