```python
# өгөгдсөн листнээс анхны тоо хайж байгаа функцыг хийе
# Анхны тоо эсэхийг шалгах функц
def is_prime(number):
    # 1 болон түүнээс бага тоо анхны тоо биш
    if number <= 1:
        return False

    # 2-оос number-ийн квадрат язгуур хүртэл шалгана
    for i in range(2, int(number ** 0.5) + 1):
        # Хэрэв үлдэгдэлгүй хуваагдаж байвал анхны тоо биш
        if number % i == 0:
            return False

    # Ямар ч тоонд үлдэгдэлгүй хуваагдаагүй бол анхны тоо
    return True


# Лист доторх анхны тоонуудыг олох функц
def find_primes(numbers):
    primes = []  # Анхны тоонуудыг хадгалах хоосон лист

    # Листийн элемент бүрийг шалгана
    for num in numbers:
        if is_prime(num):
            primes.append(num)  # Анхны тоо бол жагсаалтанд нэмнэ

    return primes


# Жишээ өгөгдөл
numbers_list = [2, 3, 4, 5, 10, 13, 17, 20, 23]

# Функцыг дуудаж үр дүнг хэвлэх
result = find_primes(numbers_list)

print("Анхны тоонууд:", result)
```

```python
# Анхны тоо эсэхийг шалгах функц
# Анхны тоо гэдэг нь 1-ээс их бөгөөд зөвхөн 1 болон өөрөөрөө үлдэгдэлгүй хуваагддаг тоог хэлнэ.
def is_prime(number):
    # 1 болон түүнээс бага тоо анхны тоо биш
    if number <= 1:
        return False
    # 2-оос number-ийн квадрат язгуур хүртэл шалгана
    for i in range(2, int(number ** 0.5) + 1):
        # Хэрэв үлдэгдэлгүй хуваагдаж байвал анхны тоо биш
        if number % i == 0:
            return False
    # Хэрэв хуваагч олдоогүй бол анхны тоо
    return True
# Лист доторх анхны тоонуудыг олох функц
def find_primes(numbers):
    primes = []  # Анхны тоонуудыг хадгалах лист
    # Листийн бүх элементийг шалгана
    for num in numbers:
        if is_prime(num):
          # append функцын төгсгөлд шинэ тоо нэмнэ
            primes.append(num)
    return primes
# Хэрэглэгчээс лист авах
# Жишээ input: 2 3 4 5 10 13
user_input = input("Тоонуудыг зайгаар тусгаарлан оруулна уу: ")
# Оруулсан string-ийг integer list болгох
# Өгөгдсөн текстийг заасан тэмдэгтээр (ихэвчлэн зайгаар) салгаж list болгоно.
numbers_list = list(map(int, user_input.split()))
# Анхны тоонуудыг олох
result = find_primes(numbers_list)
# Үр дүн хэвлэх
print("Анхны тоонууд:", result)
```

Өгөгдлийн дундаж олох функц

```python
# Өгөгдлийн дундаж олох функц
def calculate_mean(numbers):
    # Нийлбэр хадгалах хувьсагч
    total = 0
    # Листийн элемент бүрийг нэмэх
    for num in numbers:
        total += num
    # Дундаж = нийлбэр / элементийн тоо
    mean = total / len(numbers)
    return mean
# Жишээ өгөгдөл
data = [10, 20, 30, 40, 50]
# Функц дуудах
result = calculate_mean(data)
print("Дундаж:", result)
```

```python
# Өгөгдлийн дундаж олох функц
def calculate_mean(numbers):

    # Нийлбэр хадгалах хувьсагч
    total = 0

    # Листийн элемент бүрийг нэмэх
    for num in numbers:
        total += num

    # Дундаж = нийлбэр / элементийн тоо
    mean = total / len(numbers)

    return mean
# Хэрэглэгчээс өгөгдөл авах
# Жишээ: 10 20 30 40 50
user_input = input("Тоонуудыг зайгаар тусгаарлан оруулна уу: ")

# String-ийг integer list болгох
data = list(map(int, user_input.split()))

# Функц дуудах
result = calculate_mean(data)

# Үр дүн хэвлэх
print("Дундаж:", result)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Өгөгдөл үүсгэх
x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)

# Figure хэмжээ
plt.figure(figsize=(10, 5))

# Эхний график
plt.plot(
    x,
    y1,
    label="sin(x)",
    color="blue",
    linestyle="-",
    linewidth=2
)

# Хоёр дахь график
plt.plot(
    x,
    y2,
    label="cos(x)",
    color="red",
    linestyle="--",
    linewidth=2
)

# Гарчиг ба тэнхлэг
plt.title("Sin ба Cos функц")
plt.xlabel("X утга")
plt.ylabel("Y утга")

# Grid нэмэх
plt.grid(True)

# Legend
plt.legend()

# График харуулах
plt.show()
```
