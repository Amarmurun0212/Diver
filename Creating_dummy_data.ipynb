[Problem 5] Combining data

Collecting the data makes it easier to handle later. Combine the ndarrays created in Problems 1 and 4 to create an ndarray of (1000, 2). Concatenation can be done using np.concatenate()or .np.vstack()

```python
import numpy as np 
mean = [-3, 0]
cov = [[1., 0.8],[0.8, 1]]
np.random.seed(0)
x = np.random.multivariate_normal(mean, cov, (500,1))
mean1 = [0, -3]
cov1 = [[1., 0.8],[0.8, 1]]
np.random.seed(0)
z = np.random.multivariate_normal(mean1, cov1, (500,1))
concatenated_array=np.concatenate([x,z])
Answer = concatenated_array.reshape((1000,2))
print(Answer)
```
